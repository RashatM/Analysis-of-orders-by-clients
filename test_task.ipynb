```python
# очищаем фрейм от нулевых и отрицательных транзакций
transactions = transactions[transactions['m_real_amount'] > 0]
```

```python
# создаем основной фрейм данных
merge_leads = leads.merge(managers,left_on = 'l_manager_id' , right_on='manager_id', how = 'left')
merge_leads
```

# Количество заявок


```python
merge_leads['leads_count'] = merge_leads \
    .groupby(['d_utm_source','d_manager','d_club','created_at'])['lead_id'] \
    .transform('count')
merge_leads
```

# Количество мусорных заявок (на основании заявки не создан клиент)

```python
# получаем количество мусорных заявок
trash_leads_count = merge_leads[merge_leads['l_client_id'] == '00000000-0000-0000-0000-000000000000'] \
    .groupby(['d_utm_source','d_manager','d_club','created_at'], as_index = False)['lead_id'] \
    .count() \
    .rename(columns  = {'lead_id':'trash_leads_count'})

trash_leads_count
```

```python
# подтягиваем к основному фрейму
merge_leads = pd.merge(merge_leads, trash_leads_count, on = ['d_utm_source','d_manager','d_club','created_at'], how = 'left')
merge_leads
```

# Количество новых заявок (не было заявок и покупок от этого клиента раньше)


```python
# клиенты, которые не имели заявок раньше
first_clients = merge_leads \
    .groupby('l_client_id',as_index = False)['lead_id'] \
    .count() \
    .query('lead_id == 1')['l_client_id'] \
    .unique()
    
# клиенты, имеющие покупки
clients_with_transactions = transactions['l_client_id'].unique()
```

```python
# получаем количество новых заявок
new_leads_count = merge_leads \
    .query('l_client_id not in @clients_with_transactions and l_client_id  in @first_clients') \
    .groupby(['d_utm_source','d_manager','d_club','created_at'],as_index = False )['lead_id'] \
    .count() \
    .rename(columns = {'lead_id':'new_leads_count'})

new_leads_count
```

```python
# подтягиваем к основному фрейму
merge_leads = pd.merge(merge_leads, new_leads_count, on = ['d_utm_source','d_manager','d_club','created_at'], how = 'left')
merge_leads
```

# Количество покупателей (кто купил в течение недели после заявки)

```python
# подтягиваем к покупам данные о датах создания заявок
merge_clients = pd.merge(transactions, merge_leads[['lead_id',
                                                    'created_at',
                                                    'l_client_id',
                                                    'd_utm_source',
                                                    'd_manager',
                                                    'd_club']], on = 'l_client_id') \
    .rename(columns = {'created_at_x':'transactions_created_at',
                       'created_at_y':'created_at'})
merge_clients
```

```python
from datetime import  timedelta 
```

```python
# клиенты, которые купили в течение недели после заявки
clients_weekly_purchase = merge_clients[merge_clients['transactions_created_at'] \
                                .between(merge_clients['created_at'].astype('datetime64[ns]'),
                                         merge_clients['created_at'].astype('datetime64[ns]') + timedelta(weeks=1))]
```

```python
# получаем количество покупателей
clients_count = clients_weekly_purchase \
    .groupby(['d_utm_source','d_manager','d_club','created_at'], as_index = False)['l_client_id'] \
    .count() \
    .rename(columns = {'l_client_id':'clients_count'})

clients_count
```

```python
# подтягиваем к основному фрейму
merge_leads = pd.merge(merge_leads, clients_count, on = ['d_utm_source','d_manager','d_club','created_at'], how = 'left')
merge_leads
```

# Количество новых покупателей (кто купил в течение недели после заявки, и не покупал раньше)


```python
# список клиентов, у которых имеется только одна покупка, то есть не было покупок ранее
first_clients_purchase = transactions \
    .groupby('l_client_id',as_index = False)['transaction_id'] \
    .count() \
    .query('transaction_id == 1')['l_client_id'].unique()
first_clients_purchase
```

```python
# получаем количество новых клиентов
new_clients_count = clients_weekly_purchase \
    .query('l_client_id in @first_clients_purchase') \
    .groupby(['d_utm_source','d_manager','d_club','created_at'], as_index = False)['l_client_id'] \
    .count() \
    .rename(columns = {'l_client_id':'new_clients_count'})
new_clients_count
```

```python
# подтягиваем к основному фрейму
merge_leads = pd.merge(merge_leads, new_clients_count, on = ['d_utm_source','d_manager','d_club','created_at'], how = 'left')
merge_leads
```

# Доход от покупок новых покупателей

```python
# получаем доход от  новых клиентов
new_clients_income = clients_weekly_purchase \
    .query('l_client_id in @first_clients_purchase') \
    .groupby(['d_utm_source','d_manager','d_club','created_at'], as_index = False)['m_real_amount'] \
    .sum() \
    .rename(columns = {'m_real_amount':'new_clients_income'})
new_clients_income
```

```python
# подтягиваем к основному фрейму
merge_leads = pd.merge(merge_leads, new_clients_income, on = ['d_utm_source','d_manager','d_club','created_at'], how = 'left')
merge_leads
```

```python
# удаляем ненужные столбцы и дубликаты
merge_leads = merge_leads \
    .drop(columns= ['lead_id','d_utm_medium','l_manager_id', 'l_client_id', 'manager_id']) \
    .drop_duplicates(keep = 'first')
```

```python
# заменяем пустые значения
merge_leads.fillna(0,inplace = True)
```
